* В столбце `year_of_release` обнаружено 269 пропусков. Это около 1,5% от всех данных.
* Поскольку год выпуска - данные категориальные, заполнить значения медианой или средним не представляется возможным.
* Как видно из таблицы выше, в названии спортивных симуляторов часто указывается год (год в названии указывается следующий после года релиза). Для таких игр можно заполнить пропуски, взяв год из названия и уменьшить на 1.
* Также, как вариант, можно было бы найти базу данных по видеоиграм в открытых источниках и взять данные о годе выпуска из нее. К сожалению, моих знаний пока не достаточно, чтобы работать с API сайтов, предоставляющих такие данные.
* Еще один вариант, предположить год выхода игры по платформе, на которой она выходила. Например для PS2 игры выходили c 2000 по 2008-2009 года). Соответственно дата выхода игры должна быть в этом диапазоне. Однако, точную дату выхода из из этих данных выяснить невозможно. Можно было бы заполнить значения медианой, но я считаю этот вариант некорректным.


Из всех вариантов выполнимым остается только вариант, в котором год выпуска берется из имени игры.
Напишем функцию, которая заполняет такие пропуски.

```python
def fill_year(row):
    """
    Функция fill_year заполняет пропуски в данных для года выпуска игры
    на основе содержания года выхода в названии игры
    
    Принимает на вход строку таблицы с входящими в нее параметрами:
    'year_of_release' - год выхода игры
    'name' - название игры
        
    ----------
    Возвращает переменную year_of_release типа float,
    содержащую год выхода игры либо NaN, если
    в названии не оказалось указания года
    
    """
    year_of_release = row['year_of_release']
    
    if math.isnan(year_of_release):
        name = row['name']
        words = name.split(' ')
        years_list = [str(i) for i in range(1980,2017)] 
        for word in words:
            if word in years_list:
                year_of_release = float(word) - 1
            
    return year_of_release
```

Применим функцию `fill_year` ко всей таблице. После выведем пропущенные значения для проверки.

```python
data['year_of_release'] = data.apply(fill_year, axis=1)
data[data['year_of_release'].isna()]
```

Таким образом в столбце `year_of_release` удалось заполнить 15 пропусков. Это около 5% от всех пропусков в столбце `year_of_release`

### Влияние оценок на продажи

Для составления прогноза нужно учитывать все факторы, влияющие на количество продаж. Одним из таких факторов являются оценки пользователей и критиков. Рассмотрим зависимость кол-ва продаж от пользовательских оценок. Для этого построим диаграмму рассеяния и найдем коэффициенты корреляции между параметрами.

```python
fig, ax = plt.subplots(1,2,figsize=(15,6))

ax[0].set_ylim(0,100)
sns.scatterplot(data=data_by_last_5_years.query('platform == "PS3"'),
                x='total_sales',
                y='critic_score',
                alpha=0.5,
                ax=ax[0]
                )

ax[1].set_ylim(0,10)
sns.scatterplot(data=data_by_last_5_years.query('platform == "PS3"'),
                x='total_sales',
                y='user_score',
                alpha=0.5,
                ax=ax[1]
                )

corr_critic = (data_by_last_5_years.query('platform == "PS3"')['total_sales']
                .corr(data_by_last_5_years.query('platform == "PS3"')['critic_score']))
corr_user = (data_by_last_5_years.query('platform == "PS3"')['total_sales']
                .corr(data_by_last_5_years.query('platform == "PS3"')['user_score']))

fig.suptitle('Зависимость продаж от:', fontsize=16)

ax[0].set_title(f'оценок критиков \n К.корр = {corr_critic:.4}', fontsize=14)
ax[0].set_xlabel('Глобальные продажи (млн копий)', fontsize=14)
ax[0].set_ylabel('Оценка критиков', fontsize=14)

ax[1].set_title(f'оценок пользователей \n К.корр = {corr_user:.4}', fontsize=14)
ax[1].set_xlabel('Глобальные продажи (млн копий)', fontsize=14)
ax[1].set_ylabel('Оценка пользователей', fontsize=14)
```

По виду графиков и значению корреляции можно сделать следующие выводы:
* Прослеживается слабая положительная зависимость между объемами продаж и оценками критиков. Однако, нельзя с уверенностью сказать, что хорошие оценки критиков ведут к большому объему продаж. Корреляция не доказывает причинно-следственную связь. Возможно, хорошие продажи игры обеспечены высоким качеством игры, из которого и следует высокая оценка критиков. Но напрямую на продажи оценка критиков не влияет.
* Между оценками пользователей и объемами продаж взаимосвязь полностью отсутствует. В отличие от мнения критиков, мнение пользователей более субъективно. Что нравится одному - не обязательно понравится другому. Оценки пользователей никак не влияют на продажи игр.

## Проверка гипотез

### Гипотеза 1: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Сформулируем нулевую и альтернативную гипотезы:

**H0:** Средний пользовательский рейтинг игр для Xbox One равен cреднему пользовательскому рейтингу игр для PC.

$$ MO (Xbox One) = MO (PC), $$

**H1:** Средний пользовательский рейтинг игр для Xbox One не равен cреднему пользовательскому рейтингу игр для PC.

$$ MO(Xbox One) \neq MO(PC), $$

где \\(MO (Xbox One) \\) - средний пользовательский рейтинг игр для Xbox One,

\\(MO (PC) \\) - средний пользовательский рейтинг игр для PC.

Уровень значимости `alpha` примем равным `0.01`

Выведем статистические характеристики двух распределений для оценки способов решений.

```python
data_by_last_5_years.query('platform == "XOne"')['user_score'].dropna().describe()
```

```python
data_by_last_5_years.query('platform == "PC"')['user_score'].dropna().describe()
```

* Выборки представляют собой независимые наборы данных
* Выборки имеют разный размер и дисперсии
* Выбоки небольшие, что позволит нам напрямую вычислить средние и сравнить их

Для проверки гипотезы о равенстве средних будем использовать t-тест для двух независимых совокупностей, а также непосредственно сравним выборочные средние и рассмотрим распределение.

```python
alpha = 0.01

res = st.ttest_ind(data_by_last_5_years.query('platform == "XOne"')['user_score'].dropna(),
                   data_by_last_5_years.query('platform == "PC"')['user_score'].dropna(),
                   equal_var=False
                )

if res.pvalue < alpha:
    display('Отвергаем нулевую гипотезу')
else:
    display('Нулевую гипотезу отвергнуть нельзя')
    
print('P-value = ', res.pvalue)
print('Статистика разности = ', res.statistic)
```

* По результатам теста отвергнуть нулевую гипотезу нельзя.
---

Теперь построим гистограммы для двух распределений и сравним их.

```python
fig, ax = plt.subplots(1,1,figsize=(10,6))
fig.suptitle('Пользовательские рейтинги платформ Xbox One и PC', fontsize=16)

(data_by_last_5_years.query('platform == "PC"')['user_score']
                     .dropna()
                     .plot(kind='hist',
                           bins=np.arange(0,11,0.5),
                           xticks=range(0,11),
                           alpha=0.5,
                           ax=ax
                          ))

(data_by_last_5_years.query('platform == "XOne"')['user_score']
                     .dropna()
                     .plot(kind='hist',
                           bins=np.arange(0,11,0.5),
                           xticks=range(0,11),
                           alpha=0.5,
                           ax=ax
                          ))

ax.set_xlabel('Оценка пользователя', fontsize=14)
ax.set_ylabel('Кол-во оценок', fontsize=14)
ax.legend(['PC','XOne'], fontsize=14)
```

### Гипотеза 2: Средние пользовательские рейтинги жанров Action  и Sports разные

Сформулируем нулевую и альтернативную гипотезы:

**H0:** Средний пользовательский рейтинг игр жанра Action равен cреднему пользовательскому рейтингу игр жанра Sports.

$$ MO (Action) = MO (Sports), $$

**H1:** Средний пользовательский рейтинг игр жанра Action не равен cреднему пользовательскому рейтингу игр жанра Sports.

$$ MO(Action) \neq MO(Sports), $$

где \\(MO (Action) \\) - средний пользовательский рейтинг игр жанра Action,

\\(MO (Sports) \\) - средний пользовательский рейтинг игр жанра Sports.

Уровень значимости `alpha` примем равным `0.01`

Выведем статистические характеристики двух распределений для оценки способов решений.

```python
data_by_last_5_years.query('genre == "Action"')['user_score'].dropna().describe()
```

```python
data_by_last_5_years.query('genre == "Sports"')['user_score'].dropna().describe()
```

```python
alpha = 0.01

res = st.ttest_ind(data_by_last_5_years.query('genre == "Action"')['user_score'].dropna(),
                   data_by_last_5_years.query('genre == "Sports"')['user_score'].dropna(),
                   equal_var=False
                )

if res.pvalue < alpha:
    display('Отвергаем нулевую гипотезу')
else:
    display('Нулевую гипотезу отвергнуть нельзя')
    
print('P-value = ', res.pvalue)
print('Статистика разности = ', res.statistic)
```

* По результатам теста отвергаем нулевую гипотезу.
---

Теперь построим гистограммы для двух распределений и сравним их.

```python
fig, ax = plt.subplots(1,1,figsize=(10,6))
fig.suptitle('Пользовательские рейтинги платформ Xbox One и PC', fontsize=16)

(data_by_last_5_years.query('genre == "Sports"')['user_score']
                     .dropna()
                     .plot(kind='hist',
                           bins=np.arange(0,11,0.5),
                           xticks=range(0,11),
                           alpha=0.5,
                           ax=ax
                          ))

(data_by_last_5_years.query('genre == "Action"')['user_score']
                     .dropna()
                     .plot(kind='hist',
                           bins=np.arange(0,11,0.5),
                           xticks=range(0,11),
                           alpha=0.5,
                           ax=ax
                          ))

ax.set_xlabel('Оценка пользователя', fontsize=14)
ax.set_ylabel('Кол-во оценок', fontsize=14)
ax.legend(['Sports','Action'], fontsize=14)
```
