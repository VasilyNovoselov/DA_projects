### Извлечение названий улиц из адреса объекта

Еще одним важным, и, наверное, важнейшим параметром планируемого объекта общественного питания является расположение. Скорее всего не стоит распологать заведение там, где уже существует еще 10 кафе и ресторанов. Поэтому стоит проанализировать загруженность улиц объектами общественного питания. 

Для этого напишем функцию, которая будет извлекать из адреса объекта название улицы.

```python
def extract_street(df):
    
    search_res = re.search('город Москва, [0-9А-я- ёЁ№]+,', df['address']) # Ищем в адресе название улицы с городом. Пример:'город Москва, улица Пушкина,'
    search_res_poselenie = re.search('(поселение|деревня)', df['address']) # Ищем в адресе, есть ли название поселения или деревни как части города Москва. Пример:'город Москва, поселение Московское, улица Пушкина,'
    
    if search_res is None: # случай когда нет названия города, а только улицы
        df['street'] = re.search('[0-9А-я- ёЁ№]+,', df['address']).group()
        
    elif search_res_poselenie is not None: # случай когда есть подназвание 'поселение'
        df['street'] = re.search('город Москва,[0-9А-я- ёЁ№]* (поселение|деревня) [0-9А-я- ёЁ№]*, [0-9А-я- ёЁ№]*', df['address']).group()
        df['street'] = re.split(', ', df['street'])[-1]
        
    else: # в нормальном случае
        df['street'] = search_res.group()
        df['street'] = re.split(', ', df['street'])[-1][:-1]
        
    return df
```

Применим функцию к таблице с данными о ресторанах.

```python
rest_data_w_streets = rest_data.apply(extract_street, axis=1)
rest_data_w_streets.head()
```

## Геокодирование


Сами по себе названия улиц не очень информативны. Кроме как подсчитать кол-во объектов на одной улице ничего больше с ними не сделать. Поэтому я решил изучить способы получения дополнительной информации о географических объектах.

Одним из способов является геокодирование. Довольно простой библиотекой для работы с геоданными в Python является geopy. В ней есть инструменты для геокодирования и геолокации, которые помогут определить географические координаты улиц, а затем, нанести их на карту.

Использовать будем модуль геокодера - Nominatim  — инструмент для поиска данных OpenStreetMap по имени и адресу, а также для создания полных адресов точек OSM. Используем именно его, поскольку этот провайдер геоданных - бесплатен, при соблюдении ограничений. Подробнее можно прочитать на https://operations.osmfoundation.org/policies/nominatim/

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

geolocator = Nominatim(user_agent="test1")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1) # ограничитель количества запросов
```

### Определение координат улицы по адресу

Сервис Nominatim позволяет выполнять запросы на поиск данных по адресу. Сложностью для применения является то, что сервисом не рекомендуется совершать более одного запроса в секунду. Для удовлетворения этого требования выше был применен ограничитель количества запросов. Соответственно, запрос по одной улице будет выполнятся 1 секунду. Для 15600 улиц выполнение запросов займет 4,5 часа. Но поскольку улицы в таблице повторяются, количество запросов уменьшится, если записывать координаты улицы в отдельную таблицу, а потом извлекать, если соответственная улица уже геокодирована. 

Кроме этого, сервис Nominatim советует использовать кеширование данных, для избежания повторных одинаковых запросов. Напишем функцию, которая будет получать данные о координатах улицы из OSM. Кроме этого будем извлекать из полного адреса улицы название района города, это пригодится в дальнейшем. Также учтем специфику использования геокодера и предусмотрим использование кеша.

Кэш также необходимо сохранять локально, чтобы не выполнять те же самые запросы при перезапуске ядра.

```python
def get_location(df, cache):
    
    # создание столбцов с долготой, широтой и районом города
    df['lat'] = np.nan
    df['lon'] = np.nan
    df['district'] = ''
    
    # перебор каждой строки
    for i in range(len(df)):
        street = df['street'][i]
        
        # если улица не закэширована, то берем данные из геолокатора OSM
        if street not in cache['street'].values:
            location = geocode('Москва, '+ street)
            
            # если геолокатор не нашел улицу, то в таблицу записываем None
            if location is None:
                df['lat'][i] = np.nan
                df['lon'][i] = np.nan
                df['district'][i] = ''
                
            # если геолокатор нашел улицу, то записываем координаты из геолокатора
            else:
                df['lat'][i] = location.point[0]
                df['lon'][i] = location.point[1]
                district = re.search(',[0-9А-я- ёЁ№]* (район|поселение|деревня)[0-9А-я- ёЁ№]*,',location.address)
                if district is None:
                    df['district'][i] = ''
                else:
                    df['district'][i] = district.group()[2:-1]
                    
            # записываем координаты улицы в кэш, чтобы не обращатся снова к геолокатору   
            cache_row = {'street':street,
                         'district':df['district'][i],
                         'lat':df['lat'][i],
                         'lon':df['lon'][i]}
            cache = cache.append(cache_row, ignore_index=True) 
            
        # если улица уже закэширована, то берем данные о координатах из кэша
        else: 
            cached_street = cache.query('street == @street')
            df['lat'][i] = cached_street['lat']
            df['lon'][i] = cached_street['lon']
            df['district'][i] = cached_street['district'].values[0]
    
    return df, cache
    
```

Теперь создадим кэш для записи информации об улицах. Если кэш уже создан, подгрузим его.

### Определение координат районов города

Для районов также можно найти их координаты при помощи геокодирования, чтобы далее нанести информацию о них на карту. Напишем для этого функцию, подобную `get_location`.

```python
def get_district_loc(df):
    df['distr_lat'] = np.nan
    df['distr_lon'] = np.nan 
    for i in range(len(df)):
        location = geocode('Москва, '+ df['name'][i])
        if location is None:
            df['distr_lat'][i] = np.nan
            df['distr_lon'][i] = np.nan
        else:
            df['distr_lat'][i] = location.point[0]
            df['distr_lon'][i] = location.point[1]
    return df
    
```

### Зависимость числа посадочных мест от количества объектов на улице

Также стоит рассмотреть на распределение количества посадочных мест относительно числа объектов на улице. Возможно в них существует закономерность, которая поможет нам определится с расположением нашего заведения.

Построим диаграмму рассеяния, подсчитаем корреляцию и нанесем линию регрессии на график для отслеживания тренда.

```python
fig = plt.figure(figsize=(14,9))

count_number_corr = top_streets['obj_count'].corr(top_streets['number'])

sns.scatterplot(data=top_streets,
                x='obj_count',
                y='number',
                s=20,
                alpha=0.5,
                label='Улица')

sns.regplot(data=top_streets,
            x='obj_count',
            y='number',
            scatter=False,
            logx=True,
            color='red',
            label=f'Тренд, корреляция = {count_number_corr:.2f}')

plt.xlabel('Количество объектов на улице', fontsize=14)
plt.ylabel('Среднее кол-во посадочных мест для улицы', fontsize=14)
plt.title('Зависимость среднего кол-ва посадочных мест от числа заведений на улице', fontsize=18)

plt.xlim(0,200)
plt.ylim(0,300)

plt.legend(fontsize=12)

plt.show()

```

Как видно из графика, значения довольно сильно разбросаны. 
* Для улиц с 10 и менее заведениями нет характерного количества посадочных мест. Их может быть как 20, так и 120.
* В среднем на улицах с малым кол-вом заведений около 75 посадочных мест в объекте. Для улиц с большим кол-вом объектов среднее число посадочных мест около 50.
* При увеличении числа объектов на улице, разброс количества мест в заведении уменьшается.  При этом, корреляция этих двух параметров слабая и нельзя однозначно сказать о присутствии зависимости.
