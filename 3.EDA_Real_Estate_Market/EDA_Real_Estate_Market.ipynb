### Переименование столбцов

```python
data = data.rename(columns={
    'cityCenters_nearest':'city_centers_nearest_dist',
    'studio':'is_studio',
    'open_plan':'is_open_plan',
    'rooms':'room_count',
    'balcony':'balcony_count',
    'airports_nearest':'airports_nearest_dist',
    'parks_around3000':'parks_around3000_count',
    'parks_nearest':'parks_nearest_dist',
    'ponds_around3000':'ponds_around3000_count',
    'ponds_nearest':'ponds_nearest_dist',
})
data.info()
```

### Обработка пропусков

Для начала рассмотрим, сколько пропущенных значений в каждом столбце.

```python
for col in data.columns:
    if data[col].isna().sum() !=0:
        print(f'Пропущенных значений в столбце {col}:', data[col].isna().sum())
        
display(data.head(20))
```

* Пропуски в столбцах airports_nearest_dist, city_centers_nearest_dist, parks_nearest_dist и ponds_nearest_dist связаны между собой и вызваны скорее всего отсутствием координатной привязки объекта недвижимости. Из-за этого невозможно определить местоположение объекта и соответственно вычислить расстояние до интересующих объектов.
* Также если в столбцах parks_around3000_count и ponds_around3000_count	значение равно нулю, то в соответственных им столбцах с расстоянием до объекта будет стоять пропуск.
* В остальных столбцах пропуски в данных появились скорее всего просто потому что пользователь не заполнил соответствующие данные об объекте недвижимости.
* Из всех столбцов стоит заполнить пропуски только в столбце balcony_count, так как скорее всего пользователь не заполнил эту строку, потому что балкон отсутствует. Заменим пропущенные значения на нули.

```python
data['balcony_count'] = data['balcony_count'].fillna('0')
```

В столбце floors_total пропущены 86 значений. В целом, количество пропусков довольно мало, 0,3% от всех наблюдений. По-хорошему, значения количества этажей желательно было бы заполнить в соответствии с другими параметрами здания (год постройки, серия и модель здания, или другие параметры, которые могут подсказать предполагаемую высоту). Однако за неимением этих параметров остается два варианта: заполнять пропуски медианой/средним, либо не заполнять их вовсе.

В данном случае уместно заменить пропуски на медианное значение, так как количество пропусков крайне невелико.

```python
data['floors_total'] = data['floors_total'].fillna(data['floors_total'].median())
```

Теперь, следует отработать вариант, при котором заполненное медианное значения получилось меньше этажа, на котором расположена квартира. Найдем количество записей, в которых общее кол-во этажей меньше этажа квартиры.

После, заменим значение общего кол-ва этажей на текущее значение этажа.

```python
display(data.query('floors_total < floor')['floors_total'].value_counts())
data.loc[data['floors_total'] < data['floor'], ['floors_total']] = data['floor']
```

Для проверки, найдем кол-во пропусков и выведем те строки, где общее число этажей больше текущего.

```python
display(data.query('floors_total < floor'))
data['floors_total'].isna().sum()
```

Код ничего не вывел, значит пропуски были устранены.

В столбце ceiling_height присутствует очень много пропусков - 9195, что составляет почти половину всех наблюдений. Их следует заполнить. В данном случае также как и с высотой потолков хотелось бы знать год постройки и серию строения. Но поскольку этих данных у нас нет, попробуем рассчитать предполагаемую высоту потолков на основе общего кол-ва этажей в доме.

Как известно, в разное время строились дома c разными параметрами. До 1950-х годов дома строились с высокими потолками (около 3 м) и малой этажностью (до 5 этажей). C 50-х по 70-е массово строились типовые дома с наиболее распространенным кол-вом этажей - 5,9 и 14, и высотой потолков - 2,4-2,6 м. После 70-х выросло и разнообразие типовых серий домов и их планировки, высоты потолков были увеличины до 2,7 м, а этажность перестала иметь жесткое ограничение.

Благодаря этому мы сможем более точно рассчитать предпологаемые пропущенные значения, нежели методом заполнения медианой.
Для этого рассчитаем медианное значение высоты потолков для зданий разной этажности и заполним пропуски в соответствии с этажностью здания.

```python
data.groupby('floors_total')['ceiling_height'].median()
```

В таблице заметны выбивающиеся значения и пропуски в данных. Их мы обработаем позднее, а сейчас просто заполним пропуски в соответствии с медианами по этажам.

```python
ceiling_height_by_floors = data.groupby('floors_total')['ceiling_height'].transform('median')
data['ceiling_height'] = data['ceiling_height'].fillna(ceiling_height_by_floors)
```

Вспоминая о пропущенном значении в таблице для медиан, снова найдем кол-во пропусков в столбце ceiling_height

```python
data[data['ceiling_height'].isna()]
```

Остался всего один пропуск, для которого не удалось посчитать значение медианы для значения этажа. В таком случае заполним пропуск медианой высоты потолков по всем наблюдениям.

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
data['ceiling_height'].describe()
```

В столбцах living_area и kitchen_area было обнаружено около 2 тысяч пропущенных значений. Так как общая площадь квартир крайне разнообразна, простое заполнение пропусков медианой/средним будем некорректным. Более логичным способом будет определение среднего соотношения общей площади к площади жилой/кухни и заполнение данных в пропорции с общей площадью квартиры.

Для каждой записи подсчитаем отношение жилой(кухонной) площади к общей площади квартиры и выведем данные об этих столбцах.

```python
data['living_to_total_ratio'] = data['living_area']/data['total_area']
display(data['living_area'].describe())
data['living_to_total_ratio'].describe()
```

```python
data['kitchen_to_total_ratio'] = data['kitchen_area']/data['total_area']
display(data['kitchen_area'].describe())
data['kitchen_to_total_ratio'].describe()
```

В среднем, жилая площадь составляет 56,5% от площади квартиры. Площадь кухни в свою очередь составляет около 18% от общей площади квартиры. 
Теперь, зная соотношение площадей, заполним пропущенные значения, умножив общую площадь квартиры на медианное соотношение искомой площади к общей.

```python
data['living_area'] = data['living_area'].fillna(data['living_to_total_ratio'].median()*data['total_area'])
data['kitchen_area'] = data['kitchen_area'].fillna(data['kitchen_to_total_ratio'].median()*data['total_area'])
```

После этого выведем статистические данные о столбцах

```python
display(data['living_area'].describe())
data['kitchen_area'].describe()
```

Как видно из данных, все пропуски были заполнены, а такой способ заполнения не сильно сказался на статистических данных всей выборки.


```python
data.isna().sum()
```

## Расчёты и добавление результатов в таблицу

Для проведения анализа следует расчитать дополнительные параметры для каждого объекта недвижимости.

Рассчитаем для каждого объявления следующие метрики:
* Цена квадратного метра (отношение цены квартиры к ее общей площади)
* Соотношение жилой и общей площади
* Отношение площади кухни к общей

```python
data['price_per_sq_m'] = (data['last_price']/data['total_area']).round(2)
data['living_to_total_area_ratio'] = data['living_area']/data['total_area']
data['kitchen_to_total_area_ratio'] = data['kitchen_area']/data['total_area']
```

Также стоит разделить на категории значения этажа квартиры, поскольку первый и последний этаж могут влиять на стоимость.
Напишем функцию, присваивающую категорию (первый/другой/последний) этажу квартиры и добавим данные в таблицу.

```python
def floor_categorize(row):
    floor = row['floor']
    last_floor = row['floors_total']
    
    if floor == 1:
         return 'первый'
    elif floor == last_floor:
         return 'последний'
    else:
         return 'другой'

data['floor_category'] = data.apply(floor_categorize, axis=1)
```

Для более удобного сравнения данных выделим из столбца с датой объявления следующие значения:
* День недели
* Месяц
* Год публикации объявления

```python
data['exposition_weekday'] = data['first_day_exposition'].dt.weekday
data['exposition_month'] = data['first_day_exposition'].dt.month
data['exposition_year'] = data['first_day_exposition'].dt.year
data.head()
```
